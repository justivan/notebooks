```python
import pandas as pd

from smyrooms.client.distributor_client import DistributorRulesClient
```

```python
rules_client = DistributorRulesClient(
    organization_id="lgt",
    product_id="hot",
    environment="pro",
)
```

```python
def remove_items_from_list(items_to_remove, original_list):
    return [item for item in original_list if item not in items_to_remove]
```

```python
credentials_ids = ["7785", "11506"]
updated_rules = {}
expr_codes = ["EXPR", "EXPRH", "EXPRDS", "EXPRDSH"]
```

```python
for credential_id in credentials_ids:
    print(f"Processing credential ID: {credential_id}")

    response = rules_client.level_closes.get_agency_level_closes(credential_id)
    rules = response.get("rules", [])
    print(f"Found {len(rules)} total rules for credential {credential_id}")

    filters = {"tag": 1, "lvl.cli.t": 1}
    filtered_rules = rules_client.filter_rules(rules, **filters)

    # Filter out rules that have EXPR codes in lvl.prv.l when lvl.prv.t = 1 or rule_id = "1018610"
    filtered_rules = [
        rule
        for rule in filtered_rules
        if not (
            (rule.get("lvl", {}).get("prv", {}).get("t") == 1
            and any(
                code in rule.get("lvl", {}).get("prv", {}).get("l", [])
                for code in expr_codes
            ))
            or rule.get("id") == "1018610"
        )
    ]

    print(f"Filtered to {len(filtered_rules)} rules needing updates")

    for rule in filtered_rules:
        rule_id = rule["id"]

        existing_credential_ids = rule["lvl"]["cli"]["l"]
        updated_credential_ids = remove_items_from_list(
            [credential_id], existing_credential_ids
        )

        rule["lvl"]["cli"]["l"] = updated_credential_ids
        rule = rules_client.level_closes.create_or_update_level_close(rule["id"], rule)

        if rule_id not in updated_rules:
            updated_rules[rule_id] = []
        updated_rules[rule_id].append(credential_id)

        print(f"Updated rule {rule_id} - removed credential {credential_id}")

print(f"\nProcess completed. Updated {len(updated_rules)} unique rules.")
```

```python
with open("rule_updates.txt", "w") as f:
    f.write("Rule Update Results\n")
    f.write("===================\n\n")
    for rule_id, removed_credentials in updated_rules.items():
        f.write(f"Rule {rule_id:8}: Removed credentials - {removed_credentials}\n")

    f.write(f"\nTotal rules updated: {len(updated_rules)}")

print("Results written to 'rule_updates.txt'")
```
