```python
from shared.utils.backup import Backup
from shared.utils.paths import ProjectPaths
from smyrooms.client.distributor_client import DistributorRulesClient

paths = ProjectPaths("smyrooms")
paths.setup_data_dirs()

rules_client = DistributorRulesClient(
    organization_id="lgt",
    product_id="hot",
    environment="pro",
)

backup = Backup(paths.get_data_dir("backups"))
```

```python
response = rules_client.level_closes.get_level_closes()
rules = response.get("rules", [])

backup.save_to_json(rules, "level_closes")

print(f"Found {len(rules)} rules")
```
